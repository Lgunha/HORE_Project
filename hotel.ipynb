```python
# con = cx_Oracle.connect("system","1111","localhost:1521/orcl", encoding="UTF-8")
# cursor = con.cursor()

def selectAll():
    sql ="select id_resdo, detailAddr_roomNum_resdo from do_reservationtbl"
    cursor.execute(sql)
    a = cursor.fetchall()
    data  = []
    ageList=[]
    addrList=[]
    for i in a:
        age = selectAge(i[0])
        if len(age)>1:
            age=age[0:1]+'0'
        else:
            age='0'
        ageList.append(age)
        b = i[1].split("/")[0]
        bigAddr = selectAddr(b)
        addrList.append(bigAddr)


    data = pd.DataFrame(zip(addrList,ageList),columns=["지역","연령대"])
    # print(data)
    # print(data)
        #그림그리기
    #data = pd.DataFrame(data.value_counts(),columns=["cnt"])

    data.reset_index(drop=False, inplace=True)
    return data

# X = data.iloc[:,0].drop_duplicates().values
# y =data.iloc[:,1].drop_duplicates().values
# print(X,type(X))
# print(y,type(y))
# plt.scatter(X,y, color="blue")
# plt.show()
```

```python
def plotview():
    selectall = selectAll()
    uniqueAddr = selectall.iloc[:,1].unique()
    selectgroup = selectall.groupby(["지역","연령대"]).size()
    #print(dc , type(dc))
    col = 3
    row = 2 
    fig, ax = plt.subplots(ncols=col, nrows=row)
    for i,addr  in enumerate(uniqueAddr):
        data =selectgroup.loc[addr,:]
       # print(data , addr)
        x = data.index
        y = data[:]
        #print(x , y)
        sns.barplot(x=x,y=y, ax=ax[int(i/col),int(i%col)]).set(ylim=(0, 10), title=addr)
    plt.show()
```

```python
# 아이디로 나이 가져오기
def selectAge(id_resdo):
    sql = "select age_user from usertbl where id_user = " + "'"+id_resdo+"'"
    cursor.execute(sql)
    a = cursor.fetchone()
    age_user = a[0]

    return age_user

```
